```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

```python
# Step 1: Load the dataset
df = pd.read_csv('/content/Unemployment in India.csv')
print("First 5 rows of the dataset:")
df.columns = df.columns.str.strip()
print(df.head())
```

```python
# Step 3: Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
# Step 4: Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Step 5: Unemployment trend over time for a selected region (e.g., Andhra Pradesh)
plt.figure(figsize=(12, 6))
ap_data = df[df['Region'] == 'Andhra Pradesh']
sns.lineplot(data=ap_data, x='Date', y='Estimated Unemployment Rate (%)')
plt.title('Unemployment Rate Trend in Andhra Pradesh')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 9: Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Labour Metrics')
plt.tight_layout()
plt.show()
```

```python
# Step 10: Save the cleaned dataset
df.to_csv('cleaned_unemployment_data.csv', index=False)
```
